```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

path_csv = "../data_sets/caffine_content_of_drinks/caffeine.csv"
#LINK TO DATASET: https://www.kaggle.com/datasets/heitornunes/caffeine-content-of-drinks

df_drinks = pd.read_csv(path_csv)

filtered_data = df_drinks[df_drinks["Caffeine (mg)"] > 600]


sns.barplot(x="drink", y="Caffeine (mg)", data=filtered_data)
plt.xticks(rotation=90)
plt.title("Selection of caffinated drinks that exceeds overdose threshold (600 mg)") 
plt.axhline(y=600, color='r', linestyle='--', label="Caffeine Overdose Threshold (600 mg)")
plt.show()

```
